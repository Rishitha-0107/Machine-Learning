```python
import seaborn as sns
import pandas as pd
df=sns.load_dataset("tips")
print(df.head())
```

```python
#2.Understanding the dataset
df.info()
```

```python
#3.Stats Summ
df.describe()
```

```python
#4.Relationship check
#Before using any model, we must confirm relationship ..?
import matplotlib.pyplot as plt
plt.scatter(df["total_bill"],df["tip"])
plt.xlabel("Total Bill ")
plt.title("Total Bill vs Tip")
plt.show ()
```

#Observations:
1.As the bill increases tip increases
2.Relationship = Straight line (we can use linear regression)

```python
#4.Feature & Target Separation
x = df[["total_bill"]] #Feature i/p 2d only (imp)
y = df["tip"] #Feature o/p 1d
```

```python
#5.Train-Test Split
#A model must perform well on unseen data
import sklearn
from sklearn.model_selection import train_test_split
X=df[["total_bill"]]
y=df["tip"]
x_train, x_test, y_train, y_test=train_test_split(
    x,y,
    test_size=0.2,
    random_state=42
)
```

```python
# 6. Feature Standardization
# Why?
# 1. To compare coefficients fairly
# 2. To avoid dominance of large-valued features
# 3. To prepare data for Ridge and Lasso regression

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

#Imp:
1.Fit only on training data
2.Prevents the data leakage

```python
#7.Train Linear Regression Model

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train_scaled,y_train)

print("Co-efficient (m) : ", model.coef_[0])
print("Intercept (c) : " , model.intercept_ )
y_pred = model.predict(x_test_scaled)
```

```python
#8.Model Evaluation

from sklearn.metrics import r2_score

r2 = r2_score(y_test,y_pred)
print("R2 Score: " , r2)
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt (mse)

print ("MSE :", mse)
print("RMSE :" ,rmse)
```

```python
#9.Visualize the Best fit line

plt.scatter(x,y)
plt. plot(x, model.predict(scaler.transform(x)), color = "red")
plt.xlabel("Total bill")
plt.ylabel("tip")
plt.title("Linear Regression best fit line")
plt.show ()
```

```python
#10. Input - Output Real time

bill_amount = float(input("Enter the bill amount: ") )
bill_scaled =scaler.transform([[bill_amount]])
predict_tip = model.predict(bill_scaled)

print(f"Predicted Tip : {predict_tip[0] : .2f}")

```
